```python
# Problem 1: Smart Light Controller
''''Create a class to control a smart light that can be turned ON or OFF and display its status.''''
class SmartLightController:
    def __init__(self, name):
        self.name=name
        self.is_on=False

    def switch(self, action):
        if action.upper()=="ON":
            self.is_on=True
        elif action.upper()=="OFF":
            self.is_on=False
        else:
            print("Invalid Action!")

    def show_status(self):
        state="ON" if self.is_on else "OFF"
        print(f"{self.name}is{state}")

# Sample Execution
light=SmartLightController("Bedroom Light")
light.switch("ON")
light.show_status()
```

```python
# Problem 2: Employee ID Card System
'''
Create a class that stores employee name, ID, and department and displays ID card details.'''

class EmployeeIDCard:
    def __init__(self,emp_name,emp_id,dept):
        # Store employee details
        self.emp_name=emp_name
        self.emp_id=emp_id
        self.dept=dept

    def display(self):
        # Print formatted ID card
        print("\nEmployee ID Card")

        print("Name:",self.emp_name)
        print("ID:",self.emp_id)
        print("Department:",self.dept)
emp=EmployeeIDCard("Rahul","EMP102","IT")
emp.display()
```

```python
# Problem 3: Mobile Contact Record
'''Create a class to store contact name and phone number and display contact information.'''
class ContactRecord:
    def __init__(self,contact_name,phone_number):
        # Initialize contact details
        self.contact_name=contact_name
        self.phone_number=phone_number
    def save_contact(self):
        # confirmation
        print("\nContact Saved Successfully!")
        print("Name:",self.contact_name)
        print("Phone:",self.phone_number)
contact=ContactRecord("Anita","9876543210")
contact.save_contact()
```

```python
# Problem 4: Product Price Tag Generator
'''Create a class that stores product name and price and prints a formatted price tag.'''
class PriceTag:
    def __init__(self,product,amount):
        # Initialize product and price
        self.product=product
        self.amount=amount
    def print_tag(self):
        # Display formatted price tag
        print("\nPrice Tag ")
        print(f"Product:{self.product}")
        print(f"Price:₹{self.amount}")
item=PriceTag("Headphones", 2499)
item.print_tag()
```

```python
# Problem 5: Movie Rating Display System
'''Create a class that stores movie name and rating and displays movie details.'''
class MovieDetails:
    def __init__(self,title,rating):
        self.title=title
        self.rating=rating

    def show_details(self):
        # Display movie rating
        print("\nMovie Information")
        print(f"Movie:{self.title}")
        print(f"Rating:{self.rating}/5")

movie=MovieDetails("Inception",4.8)
movie.show_details()
```

```python
# Problem 6: Delivery Address Manager
'''Create a class that stores customer name and delivery address and prints delivery details.'''
class DeliveryManager:
    def __init__(self,customer,address):
        self.customer=customer
        self.address=address

    def print_delivery(self):
        # Display delivery details
        print("\nDelivery Details")
        print("Customer:",self.customer)
        print("Address:",self.address)
delivery=DeliveryManager("Suman", "Hyderabad")
delivery.print_delivery()
```
