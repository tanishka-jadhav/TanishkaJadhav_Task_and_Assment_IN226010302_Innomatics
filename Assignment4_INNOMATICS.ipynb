 Problem 1: Mobile Recharge Validation System

```python
def recharge():
    valid_plans=[199,299,399,599]

    while True:
        amount=int(input("Enter recharge amount:"))

        if amount>=50 and amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid Recharge.Try Again.")
recharge()
```

 Problem 2: Inventory Reorder Alert System

```python
def check_inventory(products):
    for item in products:
        if products[item] < 15:
            print(item,":Reorder Alert")
        else:
            print(item,":Stock OK")
inventory={
    "Pen":10, "Notebook":25,"Marker":12, "Eraser":3
    }
check_inventory(inventory)
```

 Problem 3: Student Result Processing System

```python
def check_result(marks):
    total=0
    for m in marks:
        total+=m
    average=total/len(marks)
    if average>=50:
        print("Pass")
    else:
        print("Fail")
marks=[60,45,70,55]
check_result(marks)
```

Problem 4: Cab Fare Estimator

```python
def calculate_fare():
    while True:
        km=float(input("Enter distance(km): "))
        peak=input("Peak hour?(y/n): ")
        fare=50+(12*km)
        if peak=="y":
            fare=fare+(0.25*fare)
        print("Total Fare:",fare)
        again = input("Calculate again?(y/n): ")
        if again=="n":
            break
calculate_fare()
```

Problem 5: Employee Attendance Eligibility

```python
def check_attendance(attendance):
    present=0
    for day in attendance:
        if day=="P":
            present+=1
    percentage=(present/len(attendance))*100
    if percentage>=75:
        print("Eligible")
    else:
        print("Not Eligible")
attendance = ["P","P","A","P","P","A", "P", "P"]
check_attendance(attendance)
```

Problem 6: Password Strength Checker

```python
def check_password(password):
    special="@#$"
    has_digit=False
    has_special=False

    if len(password)<8:
        print("Weak Password")
        return

    for ch in password:
        if ch.isdigit():
            has_digit=True
        if ch in special:
            has_special=True

    if has_digit and has_special:
        print("Strong Password")
    else:
        print("Weak Password")

pwd=input("Enter password: ")
check_password(pwd)
```
